```python
import pandas as pd
import matplotlib.pyplot as plt
# Example: Replace this with the raw URL of your GitHub file
github_raw_url = 'https://raw.githubusercontent.com/ranctor17/ADALL.Github/refs/heads/main/laptop_prices_2024_sgd_TL.csv'
try:
    df = pd.read_csv(github_raw_url)
    print("Successfully loaded data from GitHub!")
    display(df.head())
    display(df.info())
    display(df.describe())

    # detect missing values
    missing_values = df.isnull().sum()
    display(missing_values[missing_values > 0])

    if missing_values.sum() == 0:
        print("No missing values found in the Dataframe.")
    else:
        print("Columns with missing values are listed above.")

    # print('Missing values in each column:')
    # print(df.isnull().sum())


    # detect duplicate rows
    duplicate_rows = df[df.duplicated(keep=False)]

    print(f"Number of duplicate rows: {len(duplicate_rows)}")



    # plot each DF col graph
    # df.plot(kind='bar')

    #plot each column in df which is numeric
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
      plt.hist(df[col],bins=20)
      plt.xlabel(col)
      plt.ylabel('Frequency')
      plt.title(f'Histogram of {col}')
      plt.show()


except Exception as e:
    print(f"Error loading data: {e}")
    print("Please ensure the URL is correct and the file format is compatible with `pd.read_csv`.")

# Additional Exercise with Windsurf
# 1)What does one row represent, and what is the target?
# -identify what each row means in real life. Confirm the target variable and its type (numeric, binary, multi-class)

# 2)Is the dataset structurally sound?
# -Check number of rows and columns, data types, obvious ID columns, and whether types match their real-world meaning.

# 3)Are there missing or suspicious values?
# -Identify missing values and invalid entries. Decide whether they are meaningful, ignorable or requiring handling. If the data is already clean, prove it with check.
```
